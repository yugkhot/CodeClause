```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt


file_path = 'Iris.csv'
iris_data = pd.read_csv(file_path)


sns.pairplot(iris_data, hue='Species', diag_kind='kde')


plt.show()
```

```python
# Box plot
plt.figure(figsize=(12, 6))
sns.boxplot(x='Species', y='SepalLengthCm', data=iris_data)
plt.title('Sepal Length by Species')
plt.show()

plt.figure(figsize=(12, 6))
sns.boxplot(x='Species', y='SepalWidthCm', data=iris_data)
plt.title('Sepal Width by Species')
plt.show()

# Violin plot
plt.figure(figsize=(12, 6))
sns.violinplot(x='Species', y='PetalLengthCm', data=iris_data)
plt.title('Petal Length by Species')
plt.show()

plt.figure(figsize=(12, 6))
sns.violinplot(x='Species', y='PetalWidthCm', data=iris_data)
plt.title('Petal Width by Species')
plt.show()
```

```python
from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')

ax.scatter(iris_data['SepalLengthCm'], iris_data['SepalWidthCm'], iris_data['PetalLengthCm'],
           c=iris_data['Species'].apply(lambda x: {'Iris-setosa': 'r', 'Iris-versicolor': 'g', 'Iris-virginica': 'b'}[x]))

ax.set_xlabel('Sepal Length (cm)')
ax.set_ylabel('Sepal Width (cm)')
ax.set_zlabel('Petal Length (cm)')

plt.title('3D Scatter Plot of Sepal Length, Sepal Width, and Petal Length')
plt.show()
```
