```python
from google.colab import files
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Step 1: Upload the file
uploaded = files.upload()

# Step 2: Set the file path (the file name should match what you uploaded)
file_path = 'Iris.csv'  # Ensure this matches the file name exactly

# Load the dataset
iris_data = pd.read_csv(file_path)

# Set the aesthetic style of the plots
sns.set(style="whitegrid")

# Create a figure with subplots
fig, axes = plt.subplots(2, 2, figsize=(12, 10))

# Plot histograms for each feature
sns.histplot(iris_data['SepalLengthCm'], kde=True, ax=axes[0, 0], color='blue')
axes[0, 0].set_title('Sepal Length Distribution')

sns.histplot(iris_data['SepalWidthCm'], kde=True, ax=axes[0, 1], color='green')
axes[0, 1].set_title('Sepal Width Distribution')

sns.histplot(iris_data['PetalLengthCm'], kde=True, ax=axes[1, 0], color='red')
axes[1, 0].set_title('Petal Length Distribution')

sns.histplot(iris_data['PetalWidthCm'], kde=True, ax=axes[1, 1], color='purple')
axes[1, 1].set_title('Petal Width Distribution')

# Adjust the layout
plt.tight_layout()
plt.show()
```
